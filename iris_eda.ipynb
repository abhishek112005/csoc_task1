```python
import pandas as pd
try:
    iris_df = pd.read_csv('https://bit.ly/4nejNue')
    print(iris_df.head())
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
print("First 5 rows:\n",iris_df.head())
print("Mean: \n",iris_df.mean(numeric_only=True))
print("Median:\n",iris_df.median(numeric_only=True))
print("Standard Deviation:\n",iris_df.std(numeric_only=True))
print("All statistical Features: \n",iris_df.describe())
```

```python
#visualisation
import matplotlib.pyplot as plt
iris_df.hist(bins=10, figsize=(10, 8), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Iris Dataset Features", fontsize=16)
plt.tight_layout()
plt.show()

```
